```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("2015.csv")
```

```python
df2 = pd.read_csv("2016.csv")
```

```python
df3 = pd.read_csv("2017.csv")
```

```python
df4 = pd.read_csv("2018.csv")
```

```python
df5 = pd.read_csv("2019.csv")
```

```python
df1.info()
```

```python
df1.drop(columns=["Region", "Standard Error", "Dystopia Residual", "Happiness Rank"], inplace=True)
df1.info()
```

```python
df2.info()
```

```python
df2.drop(columns=["Region", "Lower Confidence Interval", "Upper Confidence Interval", "Dystopia Residual", "Happiness Rank"], inplace=True)
df2.info()
```

```python
df3.info()
```

```python
df3.drop(columns=["Happiness.Rank", "Whisker.high", "Whisker.low", "Dystopia.Residual"], inplace=True)

df3.columns = ['Country', 'Happiness Score', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity']
df3.info()
```

```python
df4.info()
```

```python
df4.head()
```

```python
df4.drop(columns=["Overall rank"], inplace=True)
df4.columns = ['Country', 'Happiness Score', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity']
```

```python
trust = df4["Generosity"]
gen = df4["Trust (Government Corruption)"]
df4["Trust (Government Corruption)"] = trust
df4["Generosity"] = gen
df4.head()
```

```python
df5.info()
```

```python
df5.drop(columns=["Overall rank"], inplace=True)
df5.columns = ['Country', 'Happiness Score', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity']

trust = df5["Generosity"]
gen = df5["Trust (Government Corruption)"]
df5["Trust (Government Corruption)"] = trust
df5["Generosity"] = gen
df5.head()
```

```python
data = pd.concat([df1,df2,df3,df4,df5])
data.head()
```

```python
avg_data = data.groupby(by="Country").aggregate({"Happiness Score" : np.mean, "Economy (GDP per Capita)" : np.mean, "Family" : np.mean, "Health (Life Expectancy)" : np.mean, "Freedom" : np.mean,
                                      "Trust (Government Corruption)":np.mean, "Generosity" : np.mean})
avg_data.head(10)
```

```python
# En yüksek 10 mutluluk seviyesine sahip ülke.
top_10 = avg_data.sort_values(by = "Happiness Score", ascending=False).head(10)
top_10
```

```python
fig, ax = plt.subplots()
ax.bar(top_10.index, top_10["Happiness Score"],color=(0.1, 0.1, 0.1, 0.1),  edgecolor='green')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sort_df1 = df1.sort_values(by="Country").reset_index(drop=True)
sort_df2 = df2.sort_values(by="Country").reset_index(drop=True)
sort_df3 = df3.sort_values(by="Country").reset_index(drop=True)
sort_df4 = df4.sort_values(by="Country").reset_index(drop=True)
sort_df5 = df5.sort_values(by="Country").reset_index(drop=True)
```

```python
year_happiness = pd.DataFrame()
year_happiness["Country"] = sort_df1["Country"]
year_happiness["happiness_score_2015"] = sort_df1["Happiness Score"]
year_happiness["happiness_score_2016"] = sort_df2["Happiness Score"]
year_happiness["happiness_score_2017"] = sort_df3["Happiness Score"]
year_happiness["happiness_score_2018"] = sort_df4["Happiness Score"]
year_happiness["happiness_score_2019"] = sort_df5["Happiness Score"]
year_happiness.head()
```
